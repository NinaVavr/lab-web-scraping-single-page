```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://www.popvortex.com/music/charts/top-100-songs.php'
```

```python
response = requests.get(url)
response.status_code 
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
title = []
artist = []

# define the number of iterations of our for loop
# by checking how many elements are in the retrieved result set
# (this is equivalent but more robust than just explicitly defining 250 iterations)
num_iter = len(soup.select("div.chart-content.col-xs-12.col-sm-8 > p"))

artistlist = soup.select("div.chart-content.col-xs-12.col-sm-8 > p > em")
songlist = soup.select('div.chart-content.col-xs-12.col-sm-8 > p > cite')

#iterate through the result set and retrive all the data
for i in range(num_iter):
    title.append(songlist[i].get_text())
    artist.append(artistlist[i].get_text())

print(title)
print(artist)


```
